```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import kurtosis, skew
```

```python
np.random.seed(42)

x_random = np.random.normal(loc=0, scale=2, size=10000)

plt.hist(x_random, bins='auto', edgecolor='black')
plt.title('Histogram of Random Normal Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

skewness = skew(x_random)
kurt = kurtosis(x_random)

print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurt}')
```

Skewness: The calculated skewness value should be close to 0, indicating that the distribution is approximately symmetric around the mean.
Kurtosis: The calculated kurtosis value should be close to 0, suggesting that the distribution has tails similar to those of a normal distribution (mesokurtic).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
```

```python
x = np.linspace(-5, 5, 10000)
y = 1. / (np.sqrt(2. * np.pi)) * np.exp(-.5 * (x) ** 2)

plt.hist(y, bins='auto', edgecolor='black')
plt.title('Histogram of Normal Distribution Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

skewness = skew(y)
kurt = kurtosis(y)

print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurt}')
```

Skewness: The calculated skewness value is 1.109511549276228. This indicates a right-skewed distribution, meaning there are more values on the right side of the mean.
Kurtosis: The calculated kurtosis value is -0.31039027765889804. This suggests the distribution is platykurtic, meaning it has lighter tails and a flatter peak compared to a normal distribution.
